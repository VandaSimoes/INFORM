```python
import duckdb
import plotly.express as px
import seaborn as sns
```

```python
conn = duckdb.connect()
```

```python
conn.execute("DROP TABLE IF EXISTS bupa;")
```

```python
query = """
CREATE TABLE bupa AS
SELECT
    column0 AS MCV,
    column1 AS ALP,
    column2 AS ALT,
    column3 AS AST,
    column4 AS GGT,
    column5 AS Drinks
FROM read_csv_auto('bupa.data');
"""
```

```python
conn.execute(query)
```

# Nova secção

```python
query = """
select *
from bupa
"""
```

```python
df = conn.sql(query).df()
df
```

```python
# @title Relationship between AST and ALT by Drinks

import matplotlib.pyplot as plt

# Assuming your data is in a pandas DataFrame called 'df'

plt.figure(figsize=(8, 6))
plt.scatter(df['AST'], df['ALT'], c=df['Drinks'], cmap='viridis', alpha=0.5)
plt.xlabel('AST')
plt.ylabel('ALT')
plt.title('Relationship between AST and ALT by Drinks')
_ = plt.colorbar(label='Drinks')
```

```python
bupa = conn.execute("SELECT * FROM bupa").df()
```

```python
correlation_matrix = bupa[['MCV', 'ALP', 'ALT', 'AST', 'GGT', 'Drinks']].corr()
```

```python
plt.figure(figsize=(10, 8))
```

```python
sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    square=True,
    cbar_kws={'shrink': 0.8}
)
```
